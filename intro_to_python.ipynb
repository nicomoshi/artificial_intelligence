## 2. Cell Types
IPython uses different cell types to create portable documents capable of executing code. These are typically:

1. Code,
2. Markdown,
3. Heading,
4. Raw NBConvert

The first two cell types constitute the bulk of what you would use. Headings can be accessed in a standard markdown cell, and nbconvert cells are cells that you want to convert to another format using the nbconvert command line tool.

### 2.1 Code Cells
We can make a code cell by either using the insert function, or by hitting Shift+Enter on the last cell of the notebok. If you use Jupyter Notebook, code cells are the default. If you use Google Colab, you will need to insert a code cell using the following option in the dropdown menu:

![Imgur](https://i.imgur.com/BpS4E9a.png)

The following is a code cell that your teacher will use to import some modules and write some python code. This code can be run by hitting Shift+Enter. Imported modules and variables are persisent, in the notebook's environment, which is important to keep in mind.

````python
# write code here
import numpy as np
import matplotlib.pyplot as plt

f = lambda x : x ** 2

x = np.linspace(0, 1, 1000)
y = f(x)

plt.figure()
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("This is a graph")
plt.legend(["f(x)"])
plt.show()
````

As mentioned above, imports and variables persist in the notebook. This means that if you create a variable named "xs" and later change it to "x" in your code, the variable "xs" still persists in memory. This can make it challenging to debug your code in the notebook, and is something to keep in mind. To demonstrate this persistence, we can use the imports and variables we assigned above to do operations, which we do below:

````python
g = lambda x : x ** 0.5

z = g(y)

plt.figure()
plt.plot(x, y)
plt.plot(x, z)
plt.xlabel("X")
plt.ylabel("Z")
plt.title("This is another graph")
plt.legend(["f(x)", "g(y)"])
plt.show()
````

Use the cell below to implement a function that computes $f(x) = x^3$, and plot it using the method outlined above:

````python
l = lambda x : x ** 3

i = l(y)

plt.figure()
plt.plot(x, y)
plt.plot(x, i)
plt.xlabel("X")
plt.ylabel("Z")
plt.title("This is another graph")
plt.legend(["f(x)", "g(y)"])
plt.show()
````

### 2.3 Markdown
Markdown is a way of formatting and typesetting documents using syntax that is lightweight and unobtrusive. Those of you familiar with websites like Reddit have likely encountered markdown before. Here, we'll introduce you to the basics of markdown, including creating headings, writing text, embedding images and gifs, and creating tables.

In Jupyter Notebook, you can make a cell a Markdown cell using the dropdown menu, or hitting Esc and then "M" and Enter. In Google Colab, you can make a text cell using the following icon:

![Imgur](https://i.imgur.com/zuq5LT7.png)

#### 2.3.1 Text
Text is straightforward: just type what you want into a markdown cell, and then when you run it, it will render. For example, try inputing a series of code in the below cell, and then hit Shift+Enter:

Hello

As you can see, anything we type is rendered as standard text, and we can use simple modifiers to italicize and make text bold.

#### 2.3.2 Headings
Headings are created using the # prefix, with different numbers of # being used for different levels of heading. For example, # denotes the largest heading size, ## the second largest, and so on. In code:

<code>
    # Heading 1

    ## Heading 2
    
    ### Heading 3
</code>

#### 2.3.4 Equations

Finally, we can embed equations using LATEX. In markdown:

<code>

    $$f(x) = x^2$$
</code>

This will be rendered as:

$$f(x) = x^2$$

We can use single <code>$</code> signs for rendering to a line, and <code>$$</code> for rendering to a new line. Try rendering an equation in the cell below. You can look at any latex reference to see what symbols are available.
